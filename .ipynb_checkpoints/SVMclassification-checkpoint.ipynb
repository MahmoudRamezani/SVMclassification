# SVM Classification

In the first part, a Support Vector Machine algorithm is implemented on a breast cancer dataset.  The aim of this classification is to find if a case is cancerous/Malignant or not/Benign. 
In the second part of this code, we use Principle Component Analysis (PCA) to find the first two principal components, and visualize the data in this new, two-dimensional space, with a single scatter-plot. Althogh PCA is not a classification algorithm, by using the new representation of the data, we can choose a hyperplane to classify the data in a very simple way.

```python
#importing libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC 
from sklearn.metrics import confusion_matrix,classification_report

%matplotlib inline
```

```python
#Loading datset from python (stored as a dictionary)
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
print(cancer['DESCR'])
```

```python
#Providing a pandas dataframe
df = pd.DataFrame(data=np.c_[cancer['data'], cancer['target']], columns=np.append(cancer['feature_names'],'target'))
df.head()
```

```python
# Check features, the number of data, and data type
df.info()
```

Always, the data visualization helps us to understand the features and dataset better. Here, we show how some of the features are related and corrolated with each other.

```python
sns.pairplot(df, hue='target', vars=['mean radius', 'mean texture', 'mean perimeter'])
```

```python
#To check all the correlations in numbers!
plt.figure(figsize=(20, 12))
sns.heatmap(df.corr(), annot=True)
```

```python
#split the dataset to the train and test sets
y= cancer['target']
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'], axis=1), y, test_size=0.3, random_state=101)
```

```python
# Since using SVM in the next, we have to normalize the data
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
#SVM model initialization
Model = SVC(gamma='scale')
Model.fit(X_train, y_train)
pred = Model.predict(X_test)
```

```python
# Performance Evaluation
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

According to the these resutls, the SVM performs very well here since the accuracy and precision are close to one. To make the results even better, we could think of tuning the gamma and C in SVM algorithm. To optimize these parameter, we do a grid search as follows

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001]}
grid = GridSearchCV(SVC(),param_grid,verbose=3)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

```python
# Performance Evaluation for the optimal parameters
pred_grid = grid.predict(X_test)
print(confusion_matrix(y_test,pred_grid))
print(classification_report(y_test,pred_grid))
```

```python
cm = np.array(confusion_matrix(y_test,pred_grid, labels=[1,0]))
confusion = pd.DataFrame(cm, index=['is_cancer', 'is_healthy'],
                         columns=['predicted_cancer', 'predicted_healthy'])
confusion
```

# Principle Component Analysis (PCA)

To find the principle component, first we have to normalize the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df.drop('target', axis=1))
```

```python
scaled_data = scaler.transform(df.drop('target', axis=1))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2) # the number of components we want to keep
pca.fit(scaled_data)
```

Now we can transform this data to its first 2 principal components.

```python
x_pca = pca.transform(scaled_data)
x_pca.shape
```

```python
scaled_data.shape
```

```python
df.head()
```

We've reduced 30 dimensions to just 2! Let's plot these two dimensions out!

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

Here, we can depict a straight line (or generally a hyperplan) to classify the data points in two distinct classes. Actually, the PCA is not a classification algorithm, but it can help us to transform a high dimensional dataset to a lower one and hence simplify the classification task.

```python
pca.components_
```

Each row represents a principal component, and each column corresponds to the original features. Let's visualize this relationship

```python
df_comp = pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma',)
```

In this heat map, we have shown two princeple components in the vertical axis and the initial 30 features in the horizental axis. It is clear how each feature contributes to the a principle component.
